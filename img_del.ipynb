```python
import glob
import os

img1 = sorted(glob.glob('/data/36-4/img_dir/*'))
ann1 = sorted(glob.glob('/data/36-4/ann_dir/*'))
print(len(img1))
print(len(ann1))
img2 = []
ann2 = []
```

```python
print(img1[0])
```

```python

for i in range(len(img1)):
    img_name = img1[i].split('/')[-1].split('.')[0]
    img2.append(img_name)

for j in range(len(ann1)):
    ann_name = ann1[j].split('/')[-1].split('.')[0]
    ann2.append(ann_name)

img2 = sorted(img2)
ann2 = sorted(ann2)
```

```python
sorted(ann2)
```

```python
del_list = []
for i in range(len(ann2)):
    if ann2[i] in img2:
        continue
    else:
        del_list.append(ann2[i])

print(del_list)
```

```python
print(len(del_list))
```

```python
path = '/data/36-4/ann_dir/'
for k in range(len(del_list)):
    os.remove(path+del_list[k]+'.png')
```
